```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = '2_gnssGT.csv'

time, \
gx, \
gy, \
gz = np.loadtxt(datafile, delimiter=',', unpack=True, skiprows=1)

print('Read \'%s\' successfully.' % datafile)
```

```python
fig = plt.figure(figsize=(16,12))
# Measurements
plt.plot(gx,gy, label='GNSS Measurements', c='k', lw=5)
```
